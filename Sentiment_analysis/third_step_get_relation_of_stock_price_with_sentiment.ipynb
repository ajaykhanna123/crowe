```python
df_sentiment=pd.read_csv('sentiment_apple.csv')
```

```python
df_sentiment.columns=['date','sentiment','size']
```

```python
# sentiment score selection criteria
def apply_func_total(sentiment,size):
    if sentiment==1:
        return 1*size
    elif sentiment==-1:
        return -1*size
    else:
        return 0*size
def apply_func_negative(sentiment,size):
    if sentiment==-1:
        return size
    else:
        return 0
    
def apply_func_positive(sentiment,size):
    if sentiment==1:
        return size
    else:
        return 0
    
def apply_func_neutral(sentiment,size):
    if sentiment==1:
        return size
    else:
        return 0
```

```python
df_sentiment['total_score']=df_sentiment.apply(lambda x: apply_func_total(x['sentiment'],x['size']),axis=1)
df_sentiment['negative_score']=df_sentiment.apply(lambda x: apply_func_negative(x['sentiment'],x['size']),axis=1)
df_sentiment['positive_score']=df_sentiment.apply(lambda x: apply_func_positive(x['sentiment'],x['size']),axis=1)

df_sentiment['neutral_score']=df_sentiment.apply(lambda x: apply_func_neutral(x['sentiment'],x['size']),axis=1)

```

```python
df_sentiment.head()
```

```python
df_sentiment_score=pd.DataFrame()
df_sentiment_score=df_sentiment.groupby('date')[['total_score','negative_score','positive_score','neutral_score']].sum()
```
